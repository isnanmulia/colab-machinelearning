```python
# Importing libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Preparing the data
x = [[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]]
y = [4, 5, 20, 14, 32, 22, 38, 43]
x, y = np.array(x), np.array(y)
```

```python
# Creating a model
model = LinearRegression().fit(x, y)
model
```

```python
# Getting the results
r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```

```python
# Predicting the response
y_pred = model.predict(x)
print(f"predicted response:\n{y_pred}")

y_pred2 = model.intercept_ + np.sum(model.coef_*x, axis=1)
print(f"\npredicted response:\n{y_pred2}")
```

```python
# Testing the model with new data
x_new = np.arange(10).reshape((-1, 2))
print(x_new)

y_new = model.predict(x_new)
print(f"predicted response:\n{y_new}")

y_new2 = model.intercept_ + np.sum(model.coef_*x_new, axis=1)
print(f"\npredicted response:\n{y_new2}")
```

```python
# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[:, 0], x[:, 1], y)

# Set the axis labels
ax.set_xlabel('variabel 1')
ax.set_ylabel('variabel 2')
ax.set_zlabel('respon')

# Calculate the coefficients and intercept
coef = model.coef_
intercept = model.intercept_

# Calculate the values of Z for the regression plane
xx, yy = np.meshgrid(range(2, 10), range(12, 22))
zz = coef[0]*xx + coef[1]*yy + intercept

# Add the regression plane to the 3D plot
ax.plot_surface(xx, yy, zz, alpha=0.5)

# Show the plot
plt.show()
```
