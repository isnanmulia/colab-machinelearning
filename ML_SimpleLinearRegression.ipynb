```python
# Importing libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Preparing the data
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
```

```python
# Creating a model
model = LinearRegression().fit(x, y)
model
```

```python
# Getting the results
r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```

```python
# Predicting the response
y_pred = model.predict(x)
print(f"predicted response:\n{y_pred}")

y_pred2 = model.intercept_ + model.coef_*x
print(f"\npredicted response:\n{y_pred2}")

plt.plot(x, y_pred, color='b')
plt.scatter(x, y, color='r')
plt.show()
```

```python
# Testing the model with new data
x_new = np.arange(5).reshape((-1, 1))

y_new = model.predict(x_new)
print(f"predicted response:\n{y_new}")

y_new2 = model.intercept_ + model.coef_*x_new
print(f"\npredicted response:\n{y_new2}")

plt.scatter(x_new, y_new, color='b')
plt.plot(x, y_pred, color='g')
plt.show
```
